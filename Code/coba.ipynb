```python
import math
import pandas as pd

# Konstanta Kesetimbangan Air pada 25 derajat Celcius
KW = 1e-14

def format_sci(n):
    """Format angka ke notasi ilmiah jika sangat kecil"""
    return "{:.2e}".format(n) if n < 1e-3 else "{:.3f}".format(n)
```

### SOAL

```python
# --- SOAL 1: Mencari [OH-] jika [H+] diketahui ---
h_plus_1 = 1.3
oh_minus_1 = KW / h_plus_1
print(f"1. [OH-] untuk [H+] {h_plus_1} M adalah: {format_sci(oh_minus_1)} M")
```

```python
# --- SOAL 2: Mencari [H+] jika pH diketahui ---
ph_2 = 4.82
h_plus_2 = 10**(-ph_2)
print(f"2. [H+] untuk pH {ph_2} adalah: {format_sci(h_plus_2)} M")
```

```python
# --- SOAL 3: Mencari pH jika [OH-] diketahui ---
oh_minus_3 = 2.5e-7
poh_3 = -math.log10(oh_minus_3)
ph_3 = 14 - poh_3
print(f"3. pH untuk [OH-] {oh_minus_3} M adalah: {ph_3:.3f}")
```

```python
# --- SOAL 4: pH Asam Kuat (HNO3) ---
molaritas_hno3 = 2e-3
# HNO3 adalah asam kuat valensi 1
h_plus_4 = molaritas_hno3 
ph_4 = -math.log10(h_plus_4)
print(f"4. pH larutan HNO3 {molaritas_hno3} M adalah: {ph_4:.2f}")
```

```python
# --- SOAL 5: pH Basa Kuat (Ba(OH)2) ---
molaritas_baoh2 = 1.8e-2
# Ba(OH)2 adalah basa kuat valensi 2
oh_minus_5 = 2 * molaritas_baoh2
poh_5 = -math.log10(oh_minus_5)
ph_5 = 14 - poh_5
print(f"5. pH larutan Ba(OH)2 {molaritas_baoh2} M adalah: {ph_5:.2f}")
```

### HASIL
